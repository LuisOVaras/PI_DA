# KPI 1: Aumento en el Acceso a Internet por Provincia

El KPI propuesto tiene como objetivo medir el aumento en el acceso al servicio de Internet por provincia, considerando la fórmula:

KPI = ((Nuevo acceso - Acceso actual) / Acceso actual) * 100

```python
# Especifica el año y los trimestres para comparar
ano_actual = 2020
trimestre_actual = 2
trimestre_anterior = 4 if trimestre_actual == 1 else trimestre_actual - 1

# Filtrar los datos para el año y los trimestres seleccionados
datos_actual = datos_penetracion[(datos_penetracion['Año'] == ano_actual) & (datos_penetracion['Trimestre'] == trimestre_actual)]
datos_anterior = datos_penetracion[(datos_penetracion['Año'] == ano_actual) & (datos_penetracion['Trimestre'] == trimestre_anterior)]

# Fusionar los datos actuales y anteriores en función de la provincia
datos_combinados = pd.merge(datos_actual, datos_anterior, on='Provincia', suffixes=('_actual', '_anterior'))

# Calcular el KPI
datos_combinados['KPI_Aumento_Acceso'] = ((datos_combinados['Accesos por cada 100 hogares_actual'] - datos_combinados['Accesos por cada 100 hogares_anterior']) / datos_combinados['Accesos por cada 100 hogares_anterior']) * 100

# Visualizar el KPI por provincia
kpi_por_provincia = datos_combinados[['Provincia', 'KPI_Aumento_Acceso']]
print(kpi_por_provincia)
```

```python
# Visualización del KPI 1: Aumento en el Acceso a Internet por Provincia
plt.figure(figsize=(12, 6))
sns.barplot(data=kpi_por_provincia, x='Provincia', y='KPI_Aumento_Acceso', palette='viridis')  # Puedes ajustar la paleta de colores según tus preferencias
plt.title("Aumento en el Acceso a Internet por Provincia")
plt.xlabel("Provincia")
plt.ylabel("Aumento (%)")
plt.xticks(rotation=90)
plt.tight_layout()  # Ajusta el diseño para evitar cortar etiquetas
plt.show()
```

- **Mayor aumento en el acceso a Internet**: Catamarca con un incremento del 20.76%.
- **Segundo mayor aumento**: Chubut con un incremento del 11.84%.
- **Tercer mayor aumento**: Mendoza con un incremento del 9.38%.
- **Aumento en Buenos Aires**: 1.38%.
- **Disminución en el acceso a Internet**: Santa Cruz con una disminución del -0.50%.
- **Observación general**: La mayoría de las provincias han experimentado un aumento en el acceso a Internet, indicando un progreso en la digitalización y la conectividad en Argentina.


```python
# Establecer el valor de referencia (primer trimestre disponible)
valor_referencia = datos_velocidad_media.loc[datos_velocidad_media.index[-1], 'Mbps (Media de bajada)']

# Calcular el cambio porcentual de un trimestre a otro y añadirlo como una nueva columna
datos_velocidad_media['Cambio porcentual'] = datos_velocidad_media['Mbps (Media de bajada)'].pct_change() * 100

# Calcular el KPI como el cambio porcentual acumulado desde el valor de referencia
datos_velocidad_media['KPI'] = (datos_velocidad_media['Mbps (Media de bajada)'] / valor_referencia - 1) * 100

# Mostrar los datos
print(datos_velocidad_media)
```

### KPI 2: Penetración de Internet por Trimestre (Top 5 Provincias)

##### En esta visualización, presentamos el Top 5 de provincias con la mayor penetración de Internet a lo largo de los trimestres. Esto nos permite identificar las provincias líderes en términos de acceso a Internet.

Objetivo: Mantener la penetración de Internet en las 5 provincias con la mayor penetración en un nivel no inferior al 90% durante todo el año.

Periodo de Medición: Anual

Contexto: Este KPI se centra en garantizar que las 5 provincias con la mayor penetración de Internet mantengan un nivel de acceso a Internet no inferior al 90% durante todo el año.

```python
# KPI 2: Penetración de Internet por trimestre - Top 5 provincias
# Filtrar las 5 provincias con la mayor penetración de Internet
top_provincias = datos_penetracion.groupby(['Provincia'])['Accesos por cada 100 hogares'].max().sort_values(ascending=False).head(5).index
datos_penetracion_top5 = datos_penetracion[datos_penetracion['Provincia'].isin(top_provincias)]

# Visualización de la penetración de Internet por trimestre (Top 5 provincias)
plt.figure(figsize=(10, 6))
sns.lineplot(data=datos_penetracion_top5, x='Año', y='Accesos por cada 100 hogares', hue='Provincia')
plt.title("Penetración de Internet por Trimestre (Top 5 Provincias)")
plt.xlabel("Año")
plt.ylabel("Accesos por cada 100 hogares")
plt.legend(title='Provincia', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Calcular el KPI de Penetración de Internet por Trimestre (Top 5 Provincias)
promedio_penetracion = datos_penetracion_top5.groupby(['Año'])['Accesos por cada 100 hogares'].mean()
kpi = (promedio_penetracion.mean() / 90) * 100

# Mostrar el KPI
print("KPI de Penetración de Internet por Trimestre (Top 5 Provincias):", kpi)
```

```python
# Filtrar las 5 provincias con la menor penetración de Internet
top_provincias = datos_penetracion.groupby(['Provincia'])['Accesos por cada 100 hogares'].max().sort_values(ascending=True).head(5).index
datos_penetracion_top5 = datos_penetracion[datos_penetracion['Provincia'].isin(top_provincias)]

# Visualización de la penetración de Internet por trimestre (Top 5 provincias con menor penetración)
plt.figure(figsize=(10, 6))
sns.lineplot(data=datos_penetracion_top5, x='Año', y='Accesos por cada 100 hogares', hue='Provincia')
plt.title("Penetración de Internet por Trimestre (Top 5 Provincias con Menor Penetración)")
plt.xlabel("Año")
plt.ylabel("Accesos por cada 100 hogares")
plt.legend(title='Provincia', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

```

KPI de Penetración de Internet por Trimestre (Top 5 Provincias): 85.03

Con un KPI de 85.03 para las 5 provincias con mayor penetración de Internet, podemos concluir que estas provincias han logrado mantener un nivel de acceso a Internet considerablemente alto. Sin embargo, es importante seguir monitoreando y trabajando para garantizar que esta alta penetración se mantenga estable o incluso aumente con el tiempo.
